## DATA

```python
# read web table into pandas DF
import pandas as pd

co2Link='https://docs.google.com/spreadsheets/d/e/2PACX-1vRXjfxeKHQBBCh_oHP-D6RIHEX4eduLjydHb6ZGsU4jo0IK0KKQSoYX_X1FGssC_9hnqCgjKN0K4AVf/pub?gid=775098192&single=true&output=csv'
carbon=pd.read_csv(co2Link)
```

```python
# here it is:
carbon
```

As you see, some columns have space, which should be deleted:

```python
carbon.rename(columns={'metric tonnes of CO2':'metric_tonnes_of_CO2'},inplace=True)
```

```python
#COPIA
carbon_copy=carbon.copy()
```

## PREGUNTA 1 (Keep all the columns but Ranking)

Se utiliza la funcion drop para eliminar la columna "Ranking" de manera definitiva

```python
carbon_copy.drop(columns="ranking",inplace=True)
carbon_copy
```

## PREGUNTA 2 (Find the country with the minimum CO2 emission in the world)

Como la emision es a nivel mundial se halla de manera general el pais con la menor emision de toneladas metricas de Co2

```python
carbon_copy[carbon_copy.metric_tonnes_of_CO2==min(carbon_copy.metric_tonnes_of_CO2)]
```

## PREGUNTA 3 (Find the country with the minimum CO2 emission per Region)

Se utiliza una iteracion con diferentes pasos

```python
# SE SACA LOS NOMBRES DE LAS REGIONES EN LA COLUMNA REGION
total_regiones = carbon_copy['region'].unique()

# LISTA PARA PONER LOS RESULTADOS DE LA ITERACION
resultados = []

# ITERACION
for region in total_regiones:
    # PRIMERO SE OBTIENE EL GRUPO DE PAISES DE CIERTA REGION
    datos_segun_region = carbon_copy[carbon_copy['region'] == region]

    # LUEGO SE ENCUENTRA EL PAIS CON LA MENOR EMISION DENTRO DEL GRUPO DE LA REGION
    pais_min_CO2 = datos_segun_region[datos_segun_region.metric_tonnes_of_CO2 == min(datos_segun_region.metric_tonnes_of_CO2)]

    # EN CADA ITERACION SE INSERTA CADA RESULTADO EN LA LISTA "resultados" CON LA FUNCION APPEND
    resultados.append(pais_min_CO2)

# SE CONCATENA CADA VECTOR RESULTANTE EN UN SOLO DATA FRAME CON LA FUNCION CONCAT
min_CO2_por_region = pd.concat(resultados)

# SE SELECCIONA SOLO LAS COLUMNAS PARA MOSTRAR

min_co2_por_region[['name', 'metric_tonnes_of_CO2', 'region']]
```

## PREGUNTA 4 (Create a new column by keeping the square root of the the original CO2 column)

Primero se crea la lista de valores de la columna de emision de CO2 aplicando la funcion de raiz cuadrada

```python
import math
raiz_CO2=carbon_copy.metric_tonnes_of_CO2.apply(math.sqrt)
```

Luego se aÃ±ade la columna con la funcion assingn, renombrando la nueva columna

```python
carbon_copy=carbon_copy.assign(Raiz_de_emisiones_CO2=raiz_CO2)
carbon_copy
```

## PREGUNTA 5 (Compute the average of CO2. Then, create a new column. This new column is the original CO2 minus the average computed)

Primero se halla el promedio utilizando la funcion **mean**

```python
promedio_CO2=carbon_copy.metric_tonnes_of_CO2.mean()
promedio_CO2
```

Se crea funcion para hallar la resta

```python
def resta_prom (arg):
    result=arg-promedio_CO2
    return result
```

Se aplica la funcion a la nueva columna

```python
column_resta=carbon_copy.metric_tonnes_of_CO2.apply(resta_prom)
```

Y se aÃ±ade la nueva columna

```python
carbon_copy=carbon_copy.assign(Resta_CO2_y_promedio=column_resta)
carbon_copy
```

```python
#SE GUARDA LA NUEVA TABLA
carbon_copy.to_csv("carbon_copy.csv",index=True)
```
